```python
# Dependencies
import requests
from bs4 import BeautifulSoup 
from time import sleep
import pandas as pd
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from selenium.webdriver import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#trying to scrapre price for multiple TVs at once.
#url to be scrapped
url = "https://www.harveynorman.com.au/tv-blu-ray-home-theatre/tvs-by-screen-size/all-tvs"
```

```python
#retreice page with the requests module
response = requests.get(url)
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)


browser.visit(url)
```

```python
response.text
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
#print(soup.prettify())
```

```python
#price list
prices = []
for x in range(1, 4):
    
    results = soup.find_all("span", class_ = 'price')
    
    for price in results:
        a = price.text
        prices.append(a)
        
        #print(price.text)
    
    sleep(0.1)
    browser.links.find_by_partial_text('Next')
```

```python
len(prices)
```

```python
#price list
names = []
for x in range(1, 4):
    
    results = soup.find_all("a", class_ = 'name fn l_mgn-tb-sm l_dsp-blc')
    #print(len(results))
    for price in results:
        a = price.text
        names.append(a)
        
        #print(price.text)
    
    sleep(0.1)
    browser.links.find_by_partial_text('Next')
```

```python
len(names)
```

```python
#image links
images = []
for x in range(1, 4):
    
    results = soup.find_all("img", class_ = 'photo photo-category lazy')
    print(len(results))
    for image in results:
        a = image.get("src")
        images.append(a)
        
        #print(price.text)
    
    sleep(0.1)
    browser.links.find_by_partial_text('Next')
```
