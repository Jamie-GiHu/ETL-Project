### Preliminary testing of extraction

```python
# Dependencies
import requests
from bs4 import BeautifulSoup as bs
```

```python
# URL of page to be scraped
url = 'https://www.binglee.com.au/tech/tv-video/led-lcd-tvs'
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create a Beautiful Soup object; parse with 'html.parser'
soup = bs(response.text, 'html.parser')
```

```python
# Examine the results, then determine element that contains sought info
print(soup.prettify())
```

### Identify the results wanted in the scraped HTML results

```python
# results are returned as an iterable list
results = soup.find_all('a', {"class" : "bni product-link"})
results
```
